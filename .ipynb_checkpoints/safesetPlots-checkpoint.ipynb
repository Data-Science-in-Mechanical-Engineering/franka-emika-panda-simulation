```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np
import IPython
import ipywidgets
from ipywidgets import interact, interactive, interact_manual
import os
plt.rcParams.update({'font.size': 16})
print(os.getcwd())
```

```python
options=["approximate","impedance","Full"]
control_method=ipywidgets.Dropdown(
    options=options,
    value='approximate',
    description='Control Method:',
    disabled=False,)

g2_upperbound = ipywidgets.FloatSlider(value=0.8, min=0, max=1, step=1e-2,
                                   description='Upperbound constrain 2', style={'description_width': 'initial'},
                                   continuous_update=False)
g1_lowerbound = ipywidgets.FloatSlider(value=0.05, min=0, max=1, step=1e-2,
                                   description='Lowerbound constrain 1', style={'description_width': 'initial'},
                                   continuous_update=False)


def Plot_safeset(method,g2_u,g1_l):
    fig = plt.figure(figsize=(10, 10))
    left, bottom, width, height = 0.1, 0.1, 0.8, 0.8
    ax = fig.add_axes([left, bottom, width, height])
    try:
        
        if method=="approximate":
            directory=os.getcwd()+"/approximate/arm_cost_approximate_timevarying.csv"
            data=data=pd.read_csv(directory,header=None)
        
        elif method=="impedance":
            directory=os.getcwd()+"/impedance/arm_cost_impedance_timevarying.csv"
            data=data=pd.read_csv(directory,header=None)
        
        elif method=="Full":
            directory=os.getcwd()+"/full/arm_cost_full_timevarying.csv"
            data=data=pd.read_csv(directory,header=None)
        
        data=np.asarray(data)

        colums=list([0,1,2,5,6])


        Subset_data=data[:,colums]
        data=Subset_data
        
        values = np.logical_and(data[:,2]>=g2_u,data[:,3]<=g1_l)
        
        function_values=data[:,4]
        
        function_safe=function_values[values]

        safe_max=np.where(function_values==function_safe.max())

        values=values.reshape([20,20])
        x=data[:,0]
        
        y=data[:,1]

        xmax=x[safe_max]
        ymax=y[safe_max]
        
        y = data[:, 1]

        x = x.reshape([20, 20])
        y = y.reshape([20, 20])
        values = values.reshape([20, 20])
        #y_idx = np.zeros(len(y))
        colours = ['red', 'green']
        ax.set_xlabel('q')
        ax.set_ylabel('r')
        cs = ax.contourf(x, y, values)
        ax.scatter(x, y, c=values, cmap=matplotlib.colors.ListedColormap(colours))
        ax.scatter(xmax,ymax,marker="<",color="b",s=np.asarray([200]))
        ax.set_title("Safe Set")
        ax.set_ylim([-0.1, 10.1])
        
    except KeyboardInterrupt:
        pass

    
interact_manual(Plot_safeset, method=control_method, g2_u=g2_upperbound,g1_l=g1_lowerbound);
    
```

Including input constraitns

```python
options=["approximate","impedance","Full"]
control_method=ipywidgets.Dropdown(
    options=options,
    value='impedance',
    description='Control Method:',
    disabled=False,)

g1_upperbound = ipywidgets.FloatSlider(value=0.08, min=0, max=1, step=1e-2,
                                   description='Upperbound constrain 1', style={'description_width': 'initial'},
                                   continuous_update=False)

input_upperbound = ipywidgets.FloatSlider(value=0, min=0, max=200, step=10,
                                   description='Upperbound inputs', style={'description_width': 'initial'},
                                   continuous_update=False)

vel_upperbound = ipywidgets.FloatSlider(value=7, min=0, max=8, step=1e-2,
                                   description='Upperbound velocity', style={'description_width': 'initial'},
                                   continuous_update=False)


max_eigenvalues=ipywidgets.IntSlider(value=-10, min=-20, max=0, step=1,
                                   description='Upperbound eigenvalues', style={'description_width': 'initial'},
                                   continuous_update=False)

InputConstraints_box = ipywidgets.Checkbox(False, description='Add Input constraints')

VelocityConstraint_box = ipywidgets.Checkbox(False, description='Add Velocity constraints')
def Plot_safeset(method,input_u,g1_u,vel_u,max_e,input_constraints,velocity_constraints):
    fig = plt.figure(figsize=(10, 10))
    left, bottom, width, height = 0.1, 0.1, 0.8, 0.8
    ax = fig.add_axes([left, bottom, width, height])
    try:
        if method=="approximate":
            directory=os.getcwd()+"/approximate/input_constraint/arm_cost_approximate_inputs.csv"
            data=data=pd.read_csv(directory,header=None)
        elif method=="impedance":
            directory=os.getcwd()+"/impedance/input_constraint/arm_cost_impedance_inputs3.csv"
            data=data=pd.read_csv(directory,header=None)
        
        elif method=="Full":
            directory=os.getcwd()+"/full/input_constraint/arm_cost_full_inputs.csv"
            data=data=pd.read_csv(directory,header=None)
        
               
        
        data=np.asarray(data)
        
        colums=list([0,1,5,6,7,8,9])

        
        Subset_data=data[:,colums]
        data=Subset_data
        
        values=np.ones(data.shape[0],dtype=bool)
        if input_constraints:
            if velocity_constraints:
                values = np.logical_and(data[:,4]<=input_u,data[:,5]<=vel_u)
            else:
                values=data[:,4]<=input_u
        else:
            if velocity_constraints:
                values=data[:,5]<=vel_u
            
            
        values=np.logical_and(values,data[:,2]<=g1_u)
        values=np.logical_and(values,data[:,6]<=max_e)
        print(np.sum(values))
        function_values=data[:,3]
        
        function_safe=function_values[values]

        safe_max=np.where(function_values==function_safe.max())
        shape_size=int(np.sqrt(data.shape[0]))
        values=values.reshape([shape_size,shape_size])
        x=data[:,0]
        
        y=data[:,1]

        xmax=x[safe_max]
        ymax=y[safe_max]
        
        y = data[:, 1]

        x = x.reshape([shape_size,shape_size])
        y = y.reshape([shape_size,shape_size])
        #values = values.reshape([shape_size,shape_size])
        #y_idx = np.zeros(len(y))
        colours = ['red', 'green']
        ax.set_xlabel(r'$q_c$')
        ax.set_ylabel(r'$r$')
        cs = ax.contourf(x, y, values)
        ax.scatter(x, y, c=values, cmap=matplotlib.colors.ListedColormap(colours))
        ax.scatter(xmax,ymax,marker="<",color="b",s=np.asarray([200]))
        ax.set_title("Safe Set")
        ax.set_ylim([-3.1, 3.1])
        ax.set_xlim([-6.1, 6.1])
        #plt.savefig('Safeset' + '.png', dpi=300)
        
    except KeyboardInterrupt:
        pass

    
interact_manual(Plot_safeset,method=control_method,input_u=input_upperbound,g1_u=g1_upperbound,vel_u=vel_upperbound,
                max_e=max_eigenvalues,input_constraints=InputConstraints_box,velocity_constraints=VelocityConstraint_box);
    
```

```python
def Plot_Objective(method):
    fig = plt.figure(figsize=(10, 10))
    left, bottom, width, height = 0.1, 0.1, 0.8, 0.8
    ax = fig.add_axes([left, bottom, width, height])
    try:
        if method=="approximate":
            directory=os.getcwd()+"/approximate/input_constraint/arm_cost_approximate_inputs.csv"
            data=data=pd.read_csv(directory,header=None)
        elif method=="impedance":
            directory=os.getcwd()+"/impedance/input_constraint/arm_cost_impedance_inputs2.csv"
            data=data=pd.read_csv(directory,header=None)
        
        elif method=="Full":
            directory=os.getcwd()+"/full/input_constraint/arm_cost_full_inputs.csv"
            data=data=pd.read_csv(directory,header=None)
        
               
        
        data=np.asarray(data)
        
        colums=list([0,1,5,6,7,8,9])

        
        Subset_data=data[:,colums]
        data=Subset_data
        
        
        values=data[:,3] 
        values+=0.33
        values/=0.14
        shape_size=int(np.sqrt(data.shape[0]))
        values=values.reshape([shape_size,shape_size])
        x=data[:,0]
        
        y=data[:,1]
        

        x = x.reshape([shape_size,shape_size])
        y = y.reshape([shape_size,shape_size])
        #values = values.reshape([shape_size,shape_size])
        #y_idx = np.zeros(len(y))
        colours = ['red', 'green']
        ax.set_xlabel(r'$q_c$')
        ax.set_ylabel(r'$r$')
        cs = ax.contourf(x, y, values)
        fig.colorbar(cs, ax=ax, shrink=0.9)
        #ax.scatter(x, y, c=values, cmap=matplotlib.colors.ListedColormap(colours))
        ax.set_title("Normalized Rewards")
        ax.set_ylim([-3.1, 3.1])
        ax.set_xlim([-6.1, 6.1])
        #plt.savefig('Normalized_Rewards' + '.png', dpi=300)
    except KeyboardInterrupt:
        pass

    
interact_manual(Plot_Objective,method=control_method);
```
